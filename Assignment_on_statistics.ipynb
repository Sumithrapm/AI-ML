**Calculate the mean, median, and mode for the math, reading, and writing scores**

```python
print(f"Math score mean= {df['math score'].mean()} ,Reading  Score mean={df['reading score'].mean()}, Writing Score mean={df['writing score'].mean()}",sep="....")
print(f"Math score median= {df['math score'].median()} ,Reading  Score median={df['reading score'].median()} ,Writing Score median={df['writing score'].median()}")
print(f"Math score mode= {df['math score'].mode()[0]} ,Reading  Score mode={df['reading score'].mode()[0]} ,Writing Score mode={df['writing score'].mode()[0]}") # [0] meansfirst in the list else top of list returns
```

**2. Determine the range, variance, and standard deviation for each score category**

```python
print(f"Math range= {df['math score'].max()-df['math score'].min()} varience= {df['math score'].var():.2f} std deviation= {df['math score'].std():.2f}")
print(f"Reading range= {df['reading score'].max()-df['reading score'].min()} varience= {df['reading score'].var():.2f} std deviation= {df['reading score'].std():.2f}")
print(f"Writing range= {df['writing score'].max()-df['writing score'].min()} varience= {df['writing score'].var():.2f} std deviation= {df['writing score'].std():.2f}")
```

### **Identify the mean, median, and mode of math scores for students with different parental education levels.**

```python
group_parent = df.groupby('parental level of education')['math score'].agg(['mean', 'median'])

mode_parent = df.groupby('parental level of education')['math score'].agg(lambda x: x.mode()[0]).rename('mode')

# Combine both
final_result = group_parent.join(mode_parent)

print(final_result)
```

## **Compare the mean scores of students who took the test preparation course versus those who didn’t.**

```python
compare_mean=df.groupby('test preparation course')[['math score','reading score','writing score']].mean()
print(compare_mean.round(2))
```

## **Calculate the interquartile range (IQR) for each score category.**

```python
#math IQR,outliers

Q1 = df['math score'].quantile(0.25)
Q3 = df['math score'].quantile(0.75)
IQR = Q3 - Q1
print(f"Q1= {Q1}")
print(f"Q3= {Q3}")
print(f"IQR= {IQR}")


lower = Q1 - 1.5 * IQR
upper = Q3 + 1.5 * IQR

print(f"lower= {lower}")
print(f"upper= {upper}")

outliers = df[(df['math score'] < lower) | (df['math score'] > upper)]
outliers_count = len(outliers)
print(outliers)
outliers_count
```

```python
iqr_all = df[['math score','reading score','writing score']].quantile(0.75) - \
                 df[['math score','reading score','writing score']].quantile(0.25)
print(iqr_all)
```

## **Determine the standard deviation for math scores by lunch type (free/reduced vs. standard).**

```python
std_lunch = df.groupby('lunch')['math score'].std()
print(std_lunch)
```

## **Find the correlation coefficient between math and reading scores, and then between reading and writing scores.**

```python
corr_math_read = df['math score'].corr(df['reading score'])
corr_read_write = df['reading score'].corr(df['writing score'])

print("Math vs Reading:", corr_math_read)
print("Reading vs Writing:", corr_read_write)
```


## **Interpret the strength and direction of the correlation. What might this suggest about the relationship between different scores?**

The correlation coefficient tells two things:

1. Direction (positive or negative)
2. Strength (how strong the relationship is)

 From your results:

*   Math & Reading = 0.81
Direction: Positive
Strength: Strong
 This means:

As math scores increase, reading scores also increase
Students who are good in math are generally good in reading

 * Reading & Writing = 0.95
Direction: Positive
Strength: Very Strong

This means:
Reading and writing scores move very closely together
Students strong in reading are almost always strong in writing


### **Calculate the covariance between math and reading scores. How does it compare with the correlation? What does it suggest?**

```python
cov_math_read = df['math score'].cov(df['reading score'])
print(cov_math_read)
```

Covariance → tells direction only

Correlation → tells direction + strength

👉 Covariance (186.60)

Positive value
Shows that math and reading scores increase together

Cov(X,Y)=∑(x−xˉ)(y−yˉ​)/n−1
          
👉 But:

Covariance value is not standardized
It depends on the scale of  raw data (0–100 marks)

👉 Correlation (0.81)

Also positive
Standardized value (between -1 and +1)
Clearly shows strong relationship.

Correlation=​Covariance​/σx​⋅σy

The covariance between math and reading scores is 186.60, which is positive. This indicates that both scores tend to increase together. However, covariance does not provide the strength of the relationship since it is not standardized. The correlation coefficient (0.81) confirms a strong positive relationship. This suggests that students who perform well in math are also likely to perform well in reading
