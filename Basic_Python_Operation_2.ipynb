**1. Bank Account Manager**

a. Simulate a bank account with deposit, withdrawal, and balance
enquiry operations

b. Use variables and data types to store account information

c. Implement conditional statements to check for sufficient balance
before withdrawal

d. Use functions to perform deposit, withdrawal, and balance inquiry
operations.

```python
# Bank Account Manager

# Account information
account_name = "Sumithra"
account_number = 123456
balance = 5000.0

print(account_name,type(account_name))
print(account_number,type(account_number))
print(balance,type(balance))
```

```python
# Deposit Function
def deposit(amount):
    global balance
    balance = balance + amount
    print("Amount Deposited:", amount)
    print("New Balance:", balance)


# Withdraw Function
def withdraw(amount):
    global balance
    if amount <= balance:
        balance = balance - amount
        print("Amount Withdrawn:", amount)
        print("Remaining Balance:", balance)
    else:
        print("Insufficient Balance!")


# Balance Enquiry Function
def check_balance():
    print("Account Holder:", account_name)
    print("Account Number:", account_number)
    print("Current Balance:", balance)


# Menu Driven Program
while True:
    print("\n--- Bank Account Manager ---")
    print("1. Deposit")
    print("2. Withdraw")
    print("3. Check Balance")
    print("4. Exit")

    choice = int(input("Enter your choice: "))

    if choice == 1:
        amount = float(input("Enter deposit amount: "))
        deposit(amount)

    elif choice == 2:
        amount = float(input("Enter withdrawal amount: "))
        withdraw(amount)

    elif choice == 3:
        check_balance()

    elif choice == 4:
        print("Thank you for using the Bank System")
        break

    else:
        print("Invalid Choice")
```

# **2. Weather Data Analyzer**

a. Analyze weather data (temperature, humidity, etc.) for a given region

b. Use variables and data types to store weather data

c. Implement conditional statements to determine weather conditions
(sunny, rainy, etc.)

d. Use functions to calculate average temperature and humidity, and print
results



```python
# Weather Data Analyzer


# Weather data variables
region: str = input("Enter Region Name: ")
temperature_list: list = []
humidity_list: list = []

n = int(input("Enter number of days data: "))


# Function to determine weather condition
def weather_condition(temp, humidity):
    if temp > 30 and humidity < 50:
        condition = "Sunny"
    elif humidity > 70:
        condition = "Rainy"
    elif temp < 20:
        condition = "Cold"
    else:
        condition = "Cloudy"

    print("Weather Condition:", condition)


# Function to calculate average temperature
def avg_temperature():
    avg = sum(temperature_list) / len(temperature_list)
    print("Average Temperature:", avg)


# Function to calculate average humidity
def avg_humidity():
    avg = sum(humidity_list) / len(humidity_list)
    print("Average Humidity:", avg)


# Input weather data
for i in range(n):
    print("\nEnter data for Day", i+1)

    temp = float(input("Temperature (°C): "))
    humidity = float(input("Humidity (%): "))

    temperature_list.append(temp)
    humidity_list.append(humidity)

    print("Data stored successfully")

    # Show weather condition immediately
    weather_condition(temp, humidity)


# Display results
print("\n----- Weather Analysis Report -----")
print("Region:", region)

avg_temperature()
avg_humidity()
```
